```python
def read_file(filename):
    with open(filename, encoding='utf8') as f:
        s = f.read()
    sentences = s.split('\n\n')
    words = [[line.split()[0] for line in sentence.split('\n')] for sentence in sentences]
    entities = [[line.split()[1] for line in sentence.split('\n')] for sentence in sentences]
    return words, entities    
```

```python
def delete_bio_single(entity):
    return entity.rsplit('-')[-1]

def delete_bio(entities):
    entities = [[delete_bio_single(entity) for entity in sentence] for sentence in entities]
    return entities
```

```python
def save_to_file(filename, words, entities):
    sentences = [zip(words[i], entities[i]) for i in range(len(words))]
    corpus = '\n\n'.join(['\n'.join([' '.join(word) for word in sentence]) for sentence in sentences])
    with open(filename, 'w', encoding='utf8') as f:
        f.write(corpus)
```

```python
def process_file(filename):
    words, entities = read_file('data/collection3/{}'.format(filename))
    entities = delete_bio(entities)
    save_to_file('data/collection3_2/{}'.format(filename), words, entities)
```

```python
process_file('valid.txt')
```
